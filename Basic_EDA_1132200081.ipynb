```python
ds_train, ds_test = train_test_split(dataset, test_size=0.2, random_state=0)
```

```python
ds_train.shape, ds_test.shape
```

```python
ds_corr_matrix = ds_train.corr()
```

```python
ds_corr_matrix
```

```python
ds_corr_matrix["quality"].sort_values(ascending=False)
```

We can see that out of the various attributes, alcohol shows the highest correlation with the quality

```python
ds_corr_matrix["density"].sort_values(ascending=False)
```

We can see that out of the various attributes, residual sugar shows the highest correlation with the density

```python
fig = plt.figure(figsize = (6, 4))
title = fig.suptitle("Sulphates Content in Wine", fontsize=14)
d= fig.add_subplot(1,1, 1)
d.set_xlabel("Sulphates")
d.set_ylabel("Density") 
sns.kdeplot(dataset['sulphates'], ax=d, shade=True, color='steelblue')
```

```python
plt.figure(figsize=(11,7))
sns.heatmap(dataset.corr(),cbar=True, cmap='Reds',annot=True)
```

positive correlation is represented by darker shades and negative correlation is represented by lighter shades<br>
There is a strong positive correlation of density and residual sugar and a strong negative correlation of density and alcohol
